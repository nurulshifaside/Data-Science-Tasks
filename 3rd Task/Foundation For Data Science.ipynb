```python
# Kolom Status BMI
conditions = [
    (df['BMI'] < 18.5),
    (df['BMI'] >= 18.5) & (df['BMI'] <= 24.9),
    (df['BMI'] >= 25.0) & (df['BMI'] <= 29.9),
    (df['BMI'] >= 30.0) & (df['BMI'] <= 34.9)
]
choices = ['Underweight', 'Normal', 'Overweight', 'Obesity']
df['Status BMI'] = pd.Series(pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 34.9], labels=choices))
df
```

```python
def calculate_investment(modal, tenure):
    # Checking if the input values are within the specified range
    if not (0 <= modal <= 1000000000) or not (1 <= tenure <= 25):
        print("Error: Modal atau tenure tidak sesuai ketentuan.")
        return

    # Initial investment
    total_investment = modal

    # Displaying investment for each year
    for year in range(1, tenure + 1):
        total_investment += total_investment * 0.25  # 25% fixed profit rate
        print(f"Year {year}, your total investment: {total_investment:.2f}")

# Taking input from the user
try:
    modal_investasi = int(input("Money Capital (Modal Investasi): "))
    jangka_waktu = int(input("Investment Tenure (Jangka Waktu Investasi) in years: "))

    # Calculating and displaying the investment
    calculate_investment(modal_investasi, jangka_waktu)

except ValueError:
    print("Error: Input harus berupa bilangan bulat.")
```
